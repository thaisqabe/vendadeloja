## pergunta 1 Quais regiões gastam mais dinheiro na loja?

````python
df['Region'].unique() #mostra os valores únicos
````

##as únicas regiões que aparecem são Sul, Oeste, Centro e Leste

````python
region_sales = df.groupby(['Region'])['Sales'].sum().reset_index()  #soma as vendas de cada região
top_sales = region_sales.sort_values(by='Sales', ascending=False) #ordena para que a maior apareça primeiro
print(top_sales.head(4).reset_index(drop=True))
````

#isso mostra que a região que mais gastou dinheiro na loja foi a Oeste

````python
# Criação do gráfico de barras
plt.figure(figsize=(10, 6)) #tamanho do gráfico
plt.bar(top_sales['Region'], top_sales['Sales'], color='skyblue')  #cria um gráfico de barras usando Region como rótulo e Sales como altura das barras
plt.xlabel('Region') #nome do eixo x
plt.ylabel('Total Sales') #nome do eixo y
plt.title('Regions por Sales') #título
plt.xticks(rotation=45)  # Gira os nomes das regiões para facilitar a leitura
plt.tight_layout()  # Ajusta o layout para evitar sobreposição
plt.show()
````

##esse gráfico torna mais fácil o entendimento, mostrando que a região Oeste foi a que mais gastou dinheiro na loja, seguida pela Leste, Centro e Sul.

##pergunta 2  Existe uma relação da cidade do pedido e a categoria do produto?




````python
df['City'].unique()
````

````python
df['Category'].unique()
````

````python
# Filtrar para uma categoria específica
categoria_especifica = "Furniture"
df_categoria = df[df['Category'] == categoria_especifica]

# Agrupar por cidade e contar o número de pedidos
cidade_mais_pedidos = df_categoria.groupby('City').size().reset_index(name='Furniture')

# Ordenar para encontrar a cidade com mais pedidos
cidade_mais_pedidos = cidade_mais_pedidos.sort_values(by='Furniture', ascending=False) #mostra do maior para o menor

# Exibir as 10 cidades com o maior número de pedidos dessa categoria
print(cidade_mais_pedidos.head(10))
````

````python
categoria_especifica = "Office Supplies"
df_categoria = df[df['Category'] == categoria_especifica]

cidade_mais_pedidos = df_categoria.groupby('City').size().reset_index(name='Office Supplies')

cidade_mais_pedidos = cidade_mais_pedidos.sort_values(by='Office Supplies', ascending=False)

print(cidade_mais_pedidos.head(10))
````

````python
categoria_especifica = "Technology"
df_categoria = df[df['Category'] == categoria_especifica]

cidade_mais_pedidos = df_categoria.groupby('City').size().reset_index(name='Technology')

cidade_mais_pedidos = cidade_mais_pedidos.sort_values(by='Technology', ascending=False)

print(cidade_mais_pedidos.head(10))
````

````python
import numpy as np


# Filtragem para as três categorias e contagem por cidade e categoria, seleciona essas categorias epecíficas, agrupa essas categorias e as cidades
df_categorias = df[df['Category'].isin(['Furniture', 'Office Supplies', 'Technology'])]
cidade_categorias = df_categorias.groupby(['City', 'Category']).size().unstack(fill_value=0)  #size conta o número de ocorrencias em cada grupo, unstack(fill_value=0) transforma em colunas

# Selecionar as 10 cidades com o maior número de pedidos no total
top_10_cidades = cidade_categorias.sum(axis=1).sort_values(ascending=False).head(10).index #.sum(axis=1) soma a linha #.sort_values(ascending=False) coloca o maior primeiro
cidade_categorias_top = cidade_categorias.loc[top_10_cidades]  #.loc seleciona apenas as linhas em top_10_cidades

# Configuração do gráfico de barras agrupadas
bar_width = 0.25  # Largura de cada barra
x = np.arange(len(cidade_categorias_top))  # Posições no eixo x para as cidades

# organiza para agrupar as barras, define o tamanho
plt.figure(figsize=(12, 8))
plt.bar(x - bar_width, cidade_categorias_top['Furniture'], width=bar_width, label='Furniture', color='skyblue')
plt.bar(x, cidade_categorias_top['Office Supplies'], width=bar_width, label='Office Supplies', color='salmon')
plt.bar(x + bar_width, cidade_categorias_top['Technology'], width=bar_width, label='Technology', color='lightgreen')

# Configurações de rótulos e título
plt.xlabel('City')
plt.ylabel('número de pedidos')
plt.title('Top 10 cidades pelo número de pedidos de cada categoria')
plt.xticks(x, cidade_categorias_top.index, rotation=45)
plt.legend(title='Category')
plt.tight_layout()

# Exibir o gráfico
plt.show()
````

## as informações desse gráfico mostram que a cidade que mais faz pedidos é a de Nova Iorque, sendo a categoria mais pedida por eles elementos de escritório. Também podemos observar que a categoria geral mais pedida são esses itens de escritório, seguida por móveis e tecnologia

## esses dados podem ser influenciados pelo fato de que Nova Iorque é a cidade mais populosa dos EUA, e também por isso a cidade tem muitas empresas e muitas áreas que precisam de produtos de escritório

#Pergunta 3 Qual a relação do modo de envio com o estado que o produto está sendo enviado?

````python
df['Ship Mode'].unique()
````

# Second Class
Descrição: Um método de envio que geralmente leva mais tempo.

Uso Típico: É frequentemente utilizado para encomendas que não são urgentes e onde o custo é uma preocupação. É uma escolha comum para envios padrão que não precisam de entrega rápida.

#Standard Class
Descrição: Este é o método de envio mais comum e normalmente oferece um equilíbrio entre custo e tempo de entrega. Geralmente é mais rápido que o envio de segunda classe.


#First Class
Descrição: Um método de envio mais rápido

Uso Típico: Ideal para produtos que precisam ser entregues rapidamente, mas que não são tão urgentes quanto as encomendas no mesmo dia.

#Same Day
Descrição: Este método de envio garante que a encomenda seja entregue no mesmo dia em que foi feita.


````python
df['State'].unique()
````

````python
# Encontrar o estado que mais usa cada modo de envio
estado_modo = df.groupby(['Ship Mode', 'State']).size().reset_index(name='Count')
estado_modo_max = estado_modo.loc[estado_modo.groupby('Ship Mode')['Count'].idxmax()]

# Exibir os resultados


print("Estado que mais usa cada modo de envio:")
print(estado_modo_max)
````

````python
# Filtragem para as três categorias e contagem por cidade e categoria, seleciona essas categorias epecíficas, agrupa essas categorias e as cidades
df_shipstate = df[df['Ship Mode'].isin(['First Class', 'Same Day', 'Second Class', 'Standard Class'])]
estado_envio = df_shipstate.groupby(['State', 'Ship Mode']).size().unstack(fill_value=0)  #size conta o número de ocorrencias em cada grupo, unstack(fill_value=0) transforma em colunas

# Selecionar as 10 cidades com o maior número de pedidos no total
top_10_estados = estado_envio.sum(axis=1).sort_values(ascending=False).head(10).index #.sum(axis=1) soma a linha #.sort_values(ascending=False) coloca o maior primeiro
estado_envio_top = estado_envio.loc[top_10_estados]

# Configuração do gráfico de barras agrupadas
bar_width = 0.20  # Largura de cada barra
spacing = 0.8  # Espaçamento entre os grupos
x = np.arange(len(estado_envio_top))  # Posições no eixo x para as cidades

# Plotar cada categoria com deslocamento para agrupar as barras, organiza para agrupar as barras
plt.figure(figsize=(16, 8))
plt.bar(x - 1.5 * bar_width, estado_envio_top['First Class'], width=bar_width, label='First Class', color='skyblue')
plt.bar(x - 0.5 * bar_width, estado_envio_top['Same Day'], width=bar_width, label='Same Day', color='salmon')
plt.bar(x + 0.5 * bar_width , estado_envio_top['Standard Class'], width=bar_width, label='Standard Class', color='red')
plt.bar(x + 1.5 * bar_width - spacing, estado_envio_top['Second Class'], width=bar_width, label='Second Class', color='lightgreen')


# Configurações de rótulos e título
plt.xlabel('State')
plt.ylabel('número de envios')
plt.title('Top 10 estados pelo número de envios de cada modo')
plt.xticks(x, estado_envio_top.index, rotation=45)
plt.legend(title='Ship Mode')
plt.tight_layout()

# Exibir o gráfico
plt.show()
````

## podemos observar que o modo de envio mais usado é o standard class, já que é um modo que possui um equilibrio entre o tempo de demora e o dinheiro gasto. Observamos também que a California é o estado que mais possui pedidos, e usa muito o modo standard class.  

## isso também é confirmado pelo fato de a California ser o estado mais populoso dos EUA
